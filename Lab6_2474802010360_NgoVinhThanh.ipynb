```python
# 1.2.Tích vô hướng, trực giao và ứng dụng
import numpy as np
import math
from scipy import linalg

# Khởi tạo vector
a = np.array([1, 2, -3])
b = np.array([4, 5, 6])


print(" TÍCH VÔ HƯỚNG ")
tich_inner = np.inner(a[:3], b[:3])
tich_dot = np.dot(a[:3], b[:3])

print("np.inner(a, b):", tich_inner)
print("np.dot(a, b):", tich_dot)


print("\n CHUẨN VỚI NUMPY ")
print("Chuẩn 2 (mặc định):", np.linalg.norm(a))
print("Chuẩn 1:", np.linalg.norm(a, 1))
print("Chuẩn 2:", np.linalg.norm(a, 2))
print("Chuẩn 3:", np.linalg.norm(a, 3))
print("Chuẩn 1000:", np.linalg.norm(a, 1000))


def chuan(v, k):
    return (sum(abs(i)**k for i in v))**(1/k)

print("\n CHUẨN VỚI HÀM TỰ VIẾT ")
print("chuan(a, 1):", chuan(a, 1))
print("chuan(a, 2):", chuan(a, 2))
print("chuan(a, 3):", chuan(a, 3))
print("chuan(a, 1000):", chuan(a, 1000))


print("\n CHUẨN VỚI SCIPY ")
print("scipy linalg.norm(a):", linalg.norm(a))
print("scipy linalg.norm(a, 100):", linalg.norm(a, 100))
print("scipy linalg.norm(a, 1000):", linalg.norm(a, 1000))
```

```python
# Bài tập 1
import numpy as np
import math

a = np.array([1, 2, 3])


mag_a = np.sqrt(a.dot(a))


mag_b = np.sqrt(a @ a)


mag_c = np.sqrt(np.inner(a, a))


mag = lambda x: math.sqrt(sum(i**2 for i in x))
mag_d = mag(a)

print("mag_a =", mag_a)
print("mag_b =", mag_b)
print("mag_c =", mag_c)
print("mag_d =", mag_d)
```

```python
# Bài tập 2
import numpy as np


def khoang_cach(u, v):
    return np.linalg.norm(u - v)


def goc_S(u, v):
    cos_theta = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

    cos_theta = np.clip(cos_theta, -1, 1)
    return np.arccos(cos_theta)


def phep_chieu_truc_giao(u, v):
    return (np.dot(u, v) / np.dot(v, v)) * v



u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

print("a) Khoảng cách giữa u và v là:", khoang_cach(u, v))
print("b) Góc S (radian) giữa u và v là:", goc_S(u, v))
print("c) Phép chiếu trực giao p của u trên v là:", phep_chieu_truc_giao(u, v))
```
