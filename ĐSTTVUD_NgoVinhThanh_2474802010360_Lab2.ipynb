```python
# 1. Dẫn nhập – Một số hàm về xử lý vector với Python 
def scale(a, v):
     return [a * vi for vi in v]
v = [3, 5, 7]
print(f'Kết quả scale là: {scale(10, v)}')

def sumvector(v, w):
     return [vi + wi for (vi, wi) in zip(v, w)]
v = [3, 5, 7]
w = [2, 4, 6]
print (f'Kết quả sumvector là: {sumvector(v, w)}')

def dotvector(v, w):
    return sum([vi * wi for (vi, wi) in zip(v, w)])

v = [3, 5, 7]
w = [2, 4, 6]
print(f'Kết quả dotvector là: {dotvector(v, w)}')

import math  

# Tính tích vô hướng (dot product) của 2 vector
def dotvector(v, w):
    return sum([vi * wi for (vi, wi) in zip(v, w)])

# Tính bình phương độ dài vector
def lenvector_squared(v):
    return dotvector(v, v)

# Tính độ dài vector 
def lenvector(v):
    return math.sqrt(dotvector(v, v))
w = [2, 4, 6]

print(f'Bình phương độ dài vector w là: {lenvector_squared(w)}')  
print(f'Độ dài vector w là: {lenvector(w)}')  






```

```python
# 3.2. Các phép biến đổi sơ cấp trên ma trận
import numpy as np

# 1. Xây dựng ma trận A 6x6 với các phần tử từ 0 đến 35 (reshape thành 6x6)
A = np.reshape(np.arange(36.0), (6,6))
print("Ma trận A:\n", A)
# Ma trận A gồm các số từ 0 đến 35, được sắp xếp thành 6 hàng 6 cột

# 2. Tạo ma trận đơn vị 6x6
I6 = np.identity(6)
print("Ma trận đơn vị I6:\n", I6)
# Ma trận I6 là ma trận đơn vị 6x6 với đường chéo chính bằng 1, phần còn lại bằng 0

# 3. Kích thước ma trận A (số lượng phần tử)
print("Số phần tử trong ma trận A:", A.size)  # 36

# 4. In vector đường chéo của ma trận A
print("Đường chéo của A:", np.diagonal(A))  # [ 0.  7. 14. 21. 28. 35.]

# 5. Tạo ma trận mới A = A + I6
A = A + I6
print("Ma trận A sau khi cộng với I6:\n", A)
# Mỗi phần tử trên đường chéo chính của A tăng lên 1

# 6. Tạo vector cột vecB
vecB = np.array([1., 2., 3., 4., 5., 6.])
print("Vector vecB:\n", vecB)

# 7. Tính tích ma trận A với vector vecB: C = A.dot(vecB)
C = A.dot(vecB)
print("Tích ma trận A với vecB (C = A.dot(vecB)):\n", C)
# Kết quả C là vector 6 phần tử: từng phần tử bằng tổng các tích của hàng A với vecB

# 8. Khai báo ma trận D kích thước 2x6
D = np.array([
    [1., 2., 3., 4., 5., 6.],
    [1., 0., 1., 0., 1., 0.]
])
print("Ma trận D:\n", D)

# 9. Tính tích ma trận A (6x6) với D (2x6) -> lỗi vì kích thước không phù hợp
# Lưu ý: A (6x6) nhân với D (2x6) không được, phải đổi D thành (6x2) hoặc chuyển vị D
# Giả sử ý bạn muốn A.dot(D.T) để kích thước tương thích
try:
    E = A.dot(D)
    print("Tích ma trận A với D:\n", E)
except ValueError as e:
    print("Lỗi kích thước khi nhân A.dot(D):", e)

# Sửa lại D thành kích thước 6x2 để có thể nhân với A (6x6)
F = np.array([
    [1., 1.],
    [2., 0.],
    [3., 1.],
    [4., 0.],
    [5., 1.],
    [6., 0.]
])
print("Ma trận F (6x2):\n", F)

# 10. Tính tích ma trận A (6x6) với F (6x2) -> kích thước phù hợp
G = A.dot(F)
print("Tích ma trận A với F (G = A.dot(F)):\n", G)

# 11. Tính ma trận nghịch đảo của A (nếu có)
try:
    A_inv = np.linalg.inv(A)
    print("Ma trận nghịch đảo của A:\n", A_inv)
except np.linalg.LinAlgError:
    print("Ma trận A không khả nghịch (không có ma trận nghịch đảo)")

# 12. Câu hỏi: Kết quả của lệnh np.linalg.inv(np.linalg.inv(A)) là gì?
# Trả lời: chính là ma trận A ban đầu (nghịch đảo của nghịch đảo là chính nó)
if 'A_inv' in locals():
    A_inv_inv = np.linalg.inv(A_inv)
    print("Nghịch đảo của nghịch đảo của A:\n", A_inv_inv)
    print("Kiểm tra có bằng A ban đầu không:", np.allclose(A, A_inv_inv))
```

```python
# Câu 1: Hãy sử dụng numpy để giải các phương trình sau: 
import numpy as np
from sympy import symbols, solve, integrate, apart, Matrix

# Vấn đề 1 (Problem 1): Tìm điểm giao của hai đường thẳng trong R^2
print("--- Vấn đề 1 (Problem 1): Giao điểm của hai đường thẳng trong R^2 ---")

A1 = np.array([[1, -1],
               [2,  3]])
B1 = np.array([2, 6])

try:
    solution1 = np.linalg.solve(A1, B1)
    print(f"Giải pháp cho Vấn đề 1 (x, y): {solution1}")
except np.linalg.LinAlgError as e:
    print(f"Không thể giải quyết Vấn đề 1: {e}")
print("-" * 60)

# Vấn đề 2 (Problem 2): Tìm giao điểm của ba mặt phẳng trong R^3
print("--- Vấn đề 2 (Problem 2): Giao điểm của ba mặt phẳng trong R^3 ---")


A2 = np.array([[1, -1,  0],
               [1,  1, -1],
               [1,  1,  1]])
B2 = np.array([2, 3, 6])

try:
    solution2 = np.linalg.solve(A2, B2)
    print(f"Giải pháp cho Vấn đề 2 (x, y, z): {solution2}")
except np.linalg.LinAlgError as e:
    print(f"Không thể giải quyết Vấn đề 2: {e}")
print("-" * 60)

# Vấn đề 3 (Problem 3): Tìm hệ số đa thức để đa thức thỏa các nghiệm.
print("--- Vấn đề 3 (Problem 3): Tìm hệ số của đa thức bậc hai ---")



A3 = np.array([[1, 1, 1],
               [4, 2, 1],
               [9, 3, 1]])
B3 = np.array([4, 3, 4])

try:
    coefficients = np.linalg.solve(A3, B3)
    a, b, c = coefficients
    print(f"Hệ số đa thức (a, b, c): {coefficients}")
    print(f"Đa thức: p(x) = {a:.4f}x^2 + {b:.4f}x + {c:.4f}")
except np.linalg.LinAlgError as e:
    print(f"Không thể giải quyết Vấn đề 3: {e}")
print("-" * 60)

# Vấn đề 4 (Problem 4): Tính tích phân bằng phương pháp phân rã phân thức đơn giản
print("--- Vấn đề 4 (Problem 4): Tính tích phân ---")
# Tích phân: ∫ [x(x - 3)] / [(x - 1)^2(x + 2)] dx

x = symbols('x')
integrand = (x * (x - 3)) / ((x - 1)**2 * (x + 2))

# Thực hiện phân rã phân thức đơn giản
partial_fraction = apart(integrand)
print(f"Phân rã phân thức đơn giản: {partial_fraction}")

# Tính tích phân
integral_result = integrate(partial_fraction, x)
print(f"Kết quả tích phân: {integral_result}")
print("-" * 60)
```

```python
# Câu 2: Viết các câu lệnh của sympy để giải các phương trình ở Câu 1. ---
print("\n--- Câu 2: Giải các phương trình ở Câu 1 bằng sympy ---")

# Vấn đề 1 (Problem 1) với sympy
eq1_sym = x - y - 2
eq2_sym = 2*x + 3*y - 6
sol1_sympy = solve((eq1_sym, eq2_sym), (x, y))
print(f"Giải pháp sympy cho Vấn đề 1 (x, y): {sol1_sympy}")


# Vấn đề 2 (Problem 2) với sympy
z = symbols('z')
eq3_sym = x - y - 2
eq4_sym = x + y - z - 3
eq5_sym = x + y + z - 6
sol2_sympy = solve((eq3_sym, eq4_sym, eq5_sym), (x, y, z))
print(f"Giải pháp sympy cho Vấn đề 2 (x, y, z): {sol2_sympy}")


# Vấn đề 3 (Problem 3) với sympy
a, b, c = symbols('a b c')
eq_p1 = a*(1)**2 + b*(1) + c - 4
eq_p2 = a*(2)**2 + b*(2) + c - 3
eq_p3 = a*(3)**2 + b*(3) + c - 4
sol3_sympy = solve((eq_p1, eq_p2, eq_p3), (a, b, c))
print(f"Giải pháp sympy cho Vấn đề 3 (a, b, c): {sol3_sympy}")

```

```python
# Câu 3: Tính lũy thừa bậc k của ma trận F_k và chứng minh ma trận F_k chứa 3 phần tử của Fibonacci ---
print("\n--- Câu 3: Tính lũy thừa ma trận Fibonacci và chứng minh ---")

# Khởi tạo ma trận Fibonacci cơ bản F_1
F_matrix = Matrix([[1, 1],
                   [1, 0]])

k_val = 5 # Ví dụ với k = 5, bạn có thể thay đổi giá trị này

print(f"Ma trận Fibonacci cơ bản F = \n{F_matrix}")

# Tính lũy thừa bậc k của ma trận F_matrix
F_power_k = F_matrix**k_val
print(f"\nF^{k_val} (lũy thừa bậc {k_val} của F):\n{F_power_k}")

def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        a, b = 0, 1
        for _ in range(2, n + 1):
            a, b = b, a + b
        return b


expected_F_kp1 = fibonacci(k_val + 1)
expected_F_k = fibonacci(k_val)
expected_F_km1 = fibonacci(k_val - 1) if (k_val - 1) >= 0 else 0 # F_{-1} không được định nghĩa chuẩn trong dãy này

print(f"\nGiá trị Fibonacci mong đợi cho k = {k_val}:")
print(f"F_{k_val+1} = {expected_F_kp1}")
print(f"F_{k_val} = {expected_F_k}")
print(f"F_{k_val-1} = {expected_F_km1}")

print("\nSo sánh với các phần tử của ma trận F^k:")
if F_power_k[0, 0] == expected_F_kp1:
    print(f"F^k[0, 0] ({F_power_k[0, 0]}) == F_{k_val+1} ({expected_F_kp1}) -> Đúng")
else:
    print(f"F^k[0, 0] ({F_power_k[0, 0]}) != F_{k_val+1} ({expected_F_kp1}) -> Sai")

if F_power_k[0, 1] == expected_F_k:
    print(f"F^k[0, 1] ({F_power_k[0, 1]}) == F_{k_val} ({expected_F_k}) -> Đúng")
else:
    print(f"F^k[0, 1] ({F_power_k[0, 1]}) != F_{k_val} ({expected_F_k}) -> Sai")

if F_power_k[1, 0] == expected_F_k:
    print(f"F^k[1, 0] ({F_power_k[1, 0]}) == F_{k_val} ({expected_F_k}) -> Đúng")
else:
    print(f"F^k[1, 0] ({F_power_k[1, 0]}) != F_{k_val} ({expected_F_k}) -> Sai")

if F_power_k[1, 1] == expected_F_km1:
    print(f"F^k[1, 1] ({F_power_k[1, 1]}) == F_{k_val-1} ({expected_F_km1}) -> Đúng")
else:
    print(f"F^k[1, 1] ({F_power_k[1, 1]}) != F_{k_val-1} ({expected_F_km1}) -> Sai")

print("\nChứng minh: Kết quả cho thấy các phần tử của ma trận F^k là các số Fibonacci liên tiếp:")
print(f"F_{k_val+1} (góc trên bên trái), F_{k_val} (góc trên bên phải và dưới bên trái), và F_{k_val-1} (góc dưới bên phải).")
print("Điều này xác nhận tính chất của lũy thừa ma trận Fibonacci.")

```
